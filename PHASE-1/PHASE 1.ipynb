### Import all the libraries.

### 1) Fetching the webpage Url :-   
#### ('https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1')

```python
# Opening IMDB.com page on automated browser.

url='https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'  
```

```python
# Connect to the driver.

driver=webdriver.Chrome(r"C:\Users\nehas\chromedriver.exe")
driver.maximize_window()
driver.get(url)
```

### 2) Create a Empty List of all the data to be scraped.

```python
#Creating an empty list, so that we can append the values inside it.

Sno            = []
Movies_name    = []
Duration       = []
year           = []
ratings        = []
Metascore      = []
```

### 3) Start Scraping the data using the given URL.

```python
# Extracting Sno, Movie names ,duration, years and ratings from the given URL using XPATH--->

start = 1
end   = 1770
    
for page in range(start,end):
    
    try:
        sno=driver.find_elements(By.XPATH,"//span[@class='lister-item-index unbold text-primary']")
        for i in sno:
            Sno.append(i.text)
    except NoSuchElementException as e:
        Sno.append('_')
    
# Extracting Movies_name via Xpath--->
    try:
        movie=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']/a")
        for i in movie:
            Movies_name.append(i.text)
    except NoSuchElementException as e:
        Movies_name.append('_')
    

# Extracting Duration via Xpath--->
    try:
        dur=driver.find_elements(By.XPATH,"//span[@class='runtime']")
        for i in dur:
            Duration.append(i.text)
    except NoSuchElementException as e:
        Duration.append('_')

# Extracting year via Xpath--->
    try:
        year_tag=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
        for i in year_tag:
            year.append(i.text)
    except NoSuchElementException as e:
        year.append('_')
                                    
# Extracting ratings via Xpath--->
    try:
        rating_tag=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]//strong')
        for i in rating_tag:
            ratings.append(i.text)
    except NoSuchElementException as e:
        ratings.append('_')
                                     
    try:
        next=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next.click()
    except NoSuchElementException:
        break 
```

### 4) Check the length in each-

```python
len(Sno),len(Movies_name),len(Duration),len(year),len(ratings)
```

```python
# Extracting Metascore via Xpath--->

driver.get(url)

start = 1
end = 1770

for page in range(start,end):
    try:
        met=driver.find_elements(By.XPATH,'//div[@class="ratings-bar"]')

        for i in met:
            Metascore.append(i.text)
        next = driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next.click()
    except NoSuchElementException as e:
        Metascore.append('_')
        break
        
# Removing unwanted data from metascore

meta = []

for i in range(len(Metascore)):
    if i==len(Metascore)-1772:
        break
    else:
        score = Metascore[i].strip().split("\n")
    meta.append(score)

Metascores=[]
for item in meta:
    if len(item) > 1:
        metascore1 = item[1].split(' ')[0]
        Metascores.append(metascore1)
    else:
        Metascores.append(None)

print("Metascores :",len(Metascores))
```

```python
# Extracting Director List from the given url using XPATH--->   

driver.get(url)
time.sleep(3)


director = []

start = 1
end = 1770

for page in range(start,end):
    try:
        dir = driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div/div/div[3]/p[3]')
        for i in dir:
            director.append(i.text)
        next_button = driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next_button.click()
    except NoSuchElementException:
        break

        
director_list = []

for i in director:
    dir = i.split(": ")[1].split(" |")[0]
    director_list.append(dir)

print("Directors_Name: ",len(director_list))



director_1 =[]
director_2 =[]

for name in director_list:
    names = name.split(', ')
    director_1.append(names[0])
    if len(names) > 1:
        director_2.append(names[1])
    else:
        director_2.append(None)
        
print("director1 :",len(director_1))
print("director2 :",len(director_2))
```

### 5)Creating a Dataframe-

```python
first_CSV=pd.DataFrame({"Serial No":Sno,"Movie Name":Movies_name,"Director Names":director_list,"Director_1":director_1,"Director_2":director_2,
                        "Duration":Duration,"Year":year,"Ratings":ratings,"Metascores":Metascores[0:1770]})


pd.set_option("display.max_rows",None)
print('\t' "first_CSV" '\t')
print("*"*90)
first_CSV
```

### 6)Saving the IMDB File-1 as CSV file.

```python
first_CSV.to_csv("IMDB_dataset1.csv",index=False)
```

```python
# As we have scraped the data for our first csv file let's close the webdriver.

driver.close()
```

## In second CSV-  Scraping Movies_names,Stars List, Votes, Genre, Gross collection, Popularity and Certification .

### 1) Fetching the webpage Url :-   
#### ('https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1')

```python
# Opening IMDB.com page on automated browser.

url='https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'  
```

```python
# Connect to the driver.

driver=webdriver.Chrome(r"C:\Users\nehas\chromedriver.exe")
driver.maximize_window()
driver.get(url)
```

### 2) Create a Empty List of all the data to be scraped.

```python
URL              = []
movie            = []
star1            = []
star2            = []
star3            = []
star4            = []
votes            = []
gross_collection = []
genre1           = []
genre2           = []
genre3           = []
certification    = []
popularity       = []
```

### 3) Start Scraping the data using the given URL.

```python
# Extracting URL using XPATH--->  

start = 1
end   = 1770
    
for page in range(start,end):
    link=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')
    try:
        for i in link:
            URL.append(i.get_attribute('href'))
    except NoSuchElementException as e:
        movie.append('_')
                                        
    try:
        next=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next.click()
    except NoSuchElementException:
        break         
```

```python
len(URL)
```

```python
# Extracting movie name using XPATH--->  

driver.get(url)

start = 1
end   = 1770
    
for page in range(start,end):
    movie_name = driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']/a")
    try:
        for i in movie_name:
            movie.append(i.text)
    except NoSuchElementException as e:
        movie.append('_')
        
        
# Extracting votes using XPATH--->

    vote_tag = driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]/span[2]')        
    try:
        for i in vote_tag:
            votes.append(i.text)
    except NoSuchElementException as e:
        votes.append('_')
                               
    try:
        next=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next.click()
    except NoSuchElementException:
        break         
```

```python
len(movie),len(votes)
```

```python
# Getting all the IMDB details using the above URL-->

for i in URL:
    driver.get(i)
    time.sleep(1)
    
     
# Extracting popularity by using XPATH--->
    try:
        pop = driver.find_element(By.XPATH, '//div[@class="sc-5f7fb5b4-0 cUcPIU"]')
        popularity.append(pop.text.split('\n')[0])
       
    except NoSuchElementException as e:
        popularity.append('_')
    
        
# Extracting Stars by using XPATH--->

    star = driver.find_element(By.XPATH,'//div[@class="ipc-shoveler ipc-shoveler--base ipc-shoveler--page0 title-cast__grid"]')
    s = star.text.split('\n') 
    
    if len(s)>=8:
        star1.append(s[0])
        star2.append(s[2])
        star3.append(s[4])
        star4.append(s[6])
        
    elif len(s)<8 and len(s)>=6:
        star1.append(s[0])
        star2.append(s[2])
        star3.append(s[4])
        star4.append(None)   
        
    elif len(s)<6 and len(s)>=4:
        star1.append(s[0])
        star2.append(s[2])
        star3.append(None)
        star4.append(None)   
        
    elif len(s)<4 and len(s)>=2:
        star1.append(s[0])
        star2.append(None)   
        star3.append(None)
        star4.append(None)
        
    elif len(s)<2 and len(s)!=0:
        star1.append(s[0])
        star2.append(None)
        star3.append(None)
        star4.append(None)
        
    else:
        star1.append(None)   
        star2.append(None)     
        star3.append(None)     
        star4.append(None)   

            
# Extracting genre1 using XPATH--->
    try:
        g1 = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[1]/span')
        genre1.append(g1.text)
        
    except NoSuchElementException as e:
        genre1.append(None)
        
        
# Extracting genre2 using XPATH--->
    try:
        g2 = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[2]/span')
        genre2.append(g2.text)
        
    except NoSuchElementException as e:
        genre2.append(None)
        
    
# Extracting genre3 using XPATH--->
    try:
        g3 = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[3]/span')
        genre3.append(g3.text)
        
    except NoSuchElementException as e:
        genre3.append(None)
        
        
        
# Extracting gross collection using XPATH---> 
    try:
        gc = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[12]/div[2]/ul/li[4]/div/ul/li/span')
        gross_collection.append(gc.text)
        
    except NoSuchElementException as e:
        gross_collection.append(None)
    
    
    
# Extracting certification using XPATH---->
    try:
        c = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')
        certification.append(c.text)
        
    except NoSuchElementException as e:
        certification.append(None)
          
```
