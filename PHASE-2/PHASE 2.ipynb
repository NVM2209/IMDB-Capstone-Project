#### IMPORTING NECESSARY LIBRARIES-->

```python
import sqlite3
import csv
```

#### CONNECTING TO DATABASE-->

```python
#Creating a database using connect method .

db = sqlite3.connect('IMDB_Database.db')
```

```python
#Create a cursor(temporary memory location to work).

cur= db.cursor()
```

## Creating IMDB_Table1 and IMDB_Table 2 and inserting values from CSV file into it.

###### Creating IMDB_Table1

```python
# Creating table IMDB_Table1

cur.execute('create table IMDB_Table1(Sno int primary key,Movie_name text,Director_names text,Director_1 text,Director_2 text,Duration text,Ratings float,Genre_1 text,Genre_2 text,Genre_3 text)')
db.commit()
```

```python
# Inserting values from IMDB_dataset1 CSV file into IMDB_Table1

with open ('IMDB_dataset1.csv','r')as imdbfile1:  # here imdbfile1 is the reference variable from 
    file=csv.reader(imdbfile1)                    # here csv reader will read our csv file.
    next(file)
    
    Records = 0
    for row in file:
        cur.execute("INSERT INTO IMDB_Table1(Sno, Movie_Name, Director_Names, Director_1, Director_2, Duration, Ratings,Genre_1, Genre_2, Genre_3) values(?, ?, ?, ?, ?, ?, ?,NULL, NULL, NULL)",(row[0],row[1],row[2],row[3],row[4],row[5],row[7]))
        db.commit()
        Records += 1
print("Total Records Inserted in IMDB_Table1",Records)
```

```python
# Checking the data present in IMDB_Table1

table1 = cur.execute("select * FROM IMDB_Table1")
for i in table1:
    print(i)
```

###### Creating IMDB_Table2

```python
# Creating table IMDB_Table2
cur.execute('create TABLE IMDB_Table2(Movie_name, Star1, Star2, Star3, Star4, Votes, Genre1, Genre2, Genre3, Gross_Collection, Popularity, Certification)')
```

```python
# Inserting values from IMDB_dataset2 CSV file into IMDB_Table2

with open ('IMDB_dataset2.csv','r', encoding = 'utf-8')as imdbfile2:  # here imdbfile2 is the reference variable from 
    files=csv.reader(imdbfile2)                    # here csv reader will read our csv file.
    next(files)
    
    Recordset2 = 0
    for row in files:
        cur.execute('INSERT INTO IMDB_TABLE2 VALUES(?,?,?,?,?,?,?,?,?,?,?,?)', row)
        db.commit()
        Recordset2 += 1
print("Total Records Inserted in IMDB_Table2",Recordset2)
```

```python
# Checking the data present in IMDB_Table2

table2 = cur.execute("select * FROM IMDB_Table2")
for i in table2:
    print(i)
```

#### Here we can see all values are of object data type but in some columns it contains numeric data type and we need to convert it to solve queries.

```python
# Removing commas from Votes, Popularity and Gross Collection column and converting it into numeric datatype

imdb2['Votes']= imdb2['Votes'].str.replace(',','')
imdb2['Votes']=pd.to_numeric(imdb2['Votes'],errors='coerce')                    #coerce will handle Nan values

imdb2['Popularity']=pd.to_numeric(imdb2['Popularity'],errors='coerce') 

imdb2['Gross Collection']=imdb2['Gross Collection'].str.replace(r'[\$,]','')    #removing $ and , signs from gross collection column. 
imdb2['Gross Collection']=pd.to_numeric(imdb2['Gross Collection'],errors='coerce') 
```
